```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```



1. What is the average rating for each product category?


```python
avg_ratings_by_category = df.groupby('category')['rating'].mean()

print(avg_ratings_by_category)
```

2. What are the top rating_count products by category?


```python
rating_counts = df.groupby(['category', 'product_name']).size().reset_index(name='rating_count')

sorted_rating_counts = rating_counts.sort_values(by=['category', 'rating_count'], ascending=[True, False])

top_products_by_category = sorted_rating_counts.groupby('category').head(5)

print(top_products_by_category)

```

3. What is the distribution of discounted prices vs. actual prices?


```python
df['actual_price'] = df['actual_price'].replace({'₹': '', ',': ''}, regex=True)
df['actual_price'] = df['actual_price'].astype('float')

df_melted = df.melt(value_vars=['actual_price', 'discounted_price'], var_name='Price Type', value_name='Price')

plt.figure(figsize=(12, 8))
sns.histplot(df_melted, x='Price', hue='Price Type', multiple='stack', palette='viridis', bins=30)
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

4. How does the average discount percentage vary across categories?


```python
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100

# df['discount_percentage']
avg_discount_by_category = df.groupby('category')['discount_percentage'].mean()

print(avg_discount_by_category)
```

8. What is the correlation between discounted_price and rating?


```python
df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True)
df['discounted_price'] = df['discounted_price'].astype('float')
df['discounted_price']
```

```python
df['rating'] = df['rating'].replace('|', 1)

df['rating'] = df['rating'].astype(float)
df['rating'].unique()

df['discounted_price'].unique()



correlation = df['rating'].corr(df['discounted_price'])
correlation
```

9. What are the Top 5 categories based on the highest ratings?


```python

avg_ratings = df.groupby('category')['rating'].mean()
top_categories = avg_ratings.sort_values(ascending=False).head(5)
print(top_categories)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_categories_df = top_categories.reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='rating', y='category', data=top_categories_df, palette='viridis')
plt.title('Top 5 Categories Based on Highest Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```


13. Is there any relationship between the popularity and the duration of tracks? Explore this using a scatter plot.


14. Which artist has the highest number of tracks in the dataset? Display the count of tracks for each artist using a countplot.

```python
track_counts = df2['Artist'].value_counts()

most_tracks_artist = track_counts.idxmax()
most_tracks_count = track_counts.max()

print(f"The artist with the highest number of tracks is {most_tracks_artist} with {most_tracks_count} tracks.")
```

```python

plt.figure(figsize=(12, 8))
sns.countplot(data=df2, x='Artist', order=df2['Artist'].value_counts().index, palette='viridis')
plt.title('Number of Tracks for Each Artist')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')
plt.xticks(rotation=90)  
plt.show()
```


15. What are the top 5 least popular tracks in the dataset? Provide the artist name and track name for each.

19. Does the duration of tracks vary significantly across different artists? Explore this visually using a box plot or violin plot.

```python
df = df2[['Artist', 'Duration (ms)']]

plt.figure(figsize=(12, 8))
sns.boxplot(x='Artist', y='Duration (ms)', data=df, palette='viridis')
plt.title('Distribution of Track Duration Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Track Duration')
plt.xticks(rotation=90)  
plt.show()
```


20. How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a violin plot.


```python

df = df2[['Artist', 'Popularity']]


plt.figure(figsize=(12, 8))
sns.swarmplot(x='Artist', y='Popularity', data=df, palette='viridis')
plt.title('Distribution of Track Popularity for Different Artists')
plt.xlabel('Artist')
plt.ylabel('Track Popularity')
plt.xticks(rotation=90) 
plt.show()

```

```python

plt.figure(figsize=(12, 8))
sns.violinplot(x='Artist', y='Popularity', data=df, palette='viridis')
plt.title('Distribution of Track Popularity for Different Artists')
plt.xlabel('Artist')
plt.ylabel('Track Popularity')
plt.xticks(rotation=90) 
plt.show()
```
